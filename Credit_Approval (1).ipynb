```python
data = pd.read_csv('crx.data')
```

```python
data
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
```

```python
data[['b','u','g','w','v','t','t.1','f','g.1']] = enc.fit_transform(data[['b','u','g','w','v','t','t.1','f','g.1']])
```

```python
data = data.replace('?', np.nan)
data = data.fillna(data.mean())
```

```python
data.isnull().sum()
```

```python
for cols in data:
  if data[cols].dtypes=='object':
    data = data.fillna(data[cols].value_counts().index[0])

data.isnull().sum()
```

```python
data.head()
```

```python
data.rename(columns = {'+':'Status'}, inplace = True)
```

```python
data['Status'] = data['Status'].map({'+': 1, '-': 0})
data = data.drop(columns=['00202'])
data.head()
```

```python
train = data.iloc[:,:-1]
test = data.iloc[:, -1:]
```

```python
train.columns
```

```python
test.columns
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x , train_y, test_y = train_test_split(train, test, random_state=42, train_size=0.8)
```

```python
train_x.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_x = sc.fit_transform(train_x)
test_x = sc.transform(test_x)
```

## Logisitic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(verbose=1)
lr.fit(train_x,train_y)
```

```python
pred = lr.predict(test_x)
```

```python
lr.score(test_x,test_y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(pred,test_y)
cm
```

```python
acc = accuracy_score(pred,test_y)
acc
```

## SVM

```python
from sklearn.svm import SVC
lr = SVC(verbose=1)
lr.fit(train_x,train_y)
```

```python
pred = lr.predict(test_x)
```

```python
lr.score(test_x,test_y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(pred,test_y)
cm
```

```python
acc = accuracy_score(pred,test_y)
acc
```

## SGDClassifier


```python
from sklearn.linear_model import SGDClassifier
est = SGDClassifier(l1_ratio= 0.19999999999999996, learning_rate= 'optimal', loss= 'huber', penalty= 'l2')
est.fit(train_x,train_y)
```

```python
pred = est.predict(test_x)
```

```python
est.score(test_x,test_y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(pred,test_y)
cm
```

```python
acc = accuracy_score(pred,test_y)
acc
```

## Grid Search on SGD

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import SGDClassifier
est = SGDClassifier()


parameters = {"penalty":['l1','l2'],
              "loss": ['hinge', 'log_loss', 'log', 'modified_huber', 'squared_hinge', 'perceptron', 'squared_error', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'] ,
              "l1_ratio" :np.arange(0.1,0.5,0.01),
              "learning_rate": ['constant','optimal','invscaling','adaptive']
}
grid_search = GridSearchCV(estimator = est,  
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv=10,
                           verbose=0)
grid_search.fit(train_x, train_y)
```

```python
print(grid_search.best_score_)
print(grid_search.best_params_)
print(grid_search.best_estimator_)
```

## Grid Search on Logistic Regression


```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
est = LogisticRegression()


parameters = {"penalty":['l1','l2','elasticnet','none'],
              "solver": ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'] ,
              "l1_ratio" :np.arange(0.1,0.5,0.01),
            
}
grid_search = GridSearchCV(estimator = est,  
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv=10,
                           verbose=0)
grid_search.fit(train_x, train_y)
```

```python
print(grid_search.best_score_)
print(grid_search.best_params_)
print(grid_search.best_estimator_)
```
